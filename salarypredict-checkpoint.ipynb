```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import train_test_split
```

Import Dataset and extract dependent and independent variables

```python
salary_data=pd.read_csv('Salary_Data.csv')
X = salary_data.iloc[:, :-1].values
y = salary_data.iloc[:, :1].values
```

```python
salary_data
```

Visualising the Dataset

```python
sns.distplot(salary_data['YearsExperience'],kde=False,bins=10)
```

```python
sns.countplot(y='YearsExperience',data=salary_data)
```

```python
sns.barplot(x='YearsExperience',y='Salary',data=salary_data)
```

```python
sns.heatmap(salary_data.corr())
```

Spilitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state = 0)
```

Fitting Simple Linear Regression to the Trainig set

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train, y_train)
```

Predicting the Test set results

```python
y_pred = lr.predict(X_test)
y_pred
```

Visualising the Trainig set result

```python
plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.title('Salary ~ Experience (Train set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Visualising the Test set result

```python
plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_test, lr.predict(X_test), color = 'red')
plt.title('Salary vs Experience (Train set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Finding the Residuals

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test,y_pred))
print('MSE:', metrics.mean_squared_error(y_test,y_pred))
print('RMSE:', np.sqrt(metrics.mean_absolute_error(y_test,y_pred)))
```
